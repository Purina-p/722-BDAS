```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("Excel Processing with Spark") \
    .config("spark.jars.packages", "com.crealytics:spark-excel_2.12:0.13.7") \
    .getOrCreate()
```

```python
file_name = 'raw_data.xlsx'

df_spark = spark.read.format("com.crealytics.spark.excel") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load(file_name)

df_spark.printSchema()
df_spark.show()
```

```python
df_spark.describe().show()
```

```python
df = df_spark.toPandas()
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Heatmap
pivot_df = df.pivot('date', 'location', 'total_deaths')
selected_cities = ['United States', 'Brazil', 'India']
sns.heatmap(pivot_df[selected_cities], cmap='YlGnBu')
plt.title('Total Deaths Over Time for Selected Cities')
plt.show()
```

```python
# Histogram for GDP per Capita
plt.figure(figsize=(10,6))
df['gdp_per_capita'].dropna().hist(bins=50)
plt.title('Distribution of GDP per Capita')
plt.xlabel('GDP per Capita')
plt.ylabel('Frequency')
plt.show()
```

```python
# Histogram for Human Development Index
plt.figure(figsize=(10,6))
df['human_development_index'].dropna().hist(bins=50)
plt.title('Distribution of Human Development Index')
plt.xlabel('Human Development Index')
plt.ylabel('Frequency')
plt.show()
```

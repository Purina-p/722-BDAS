```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree               
from sklearn.tree import export_text             
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.api as sm
```

```python
file_path = "Updated_Merged_Dataset.csv"
merged_df = pd.read_csv(file_path)
```

```python
selected_features = ['HDI', 'TC', 'TD', 'STI', 'POP']

df = merged_df[selected_features + ['GDPCAP']].copy()

if not df['GDPCAP'].dtype == 'O':  
    threshold = df['GDPCAP'].median()
    df['GDPCAP'] = ['high' if x >= threshold else 'low' for x in df['GDPCAP']]

X = df[selected_features]
y = df['GDPCAP']



clf = DecisionTreeClassifier(max_depth=5)  
clf = clf.fit(X, y)

print("\nFeature Importance from CR Tree:")
for feature, importance in zip(selected_features, clf.feature_importances_):
    print(f"{feature}: {importance}")

tree_rules = export_text(clf, feature_names=list(X.columns))
print(tree_rules)

plt.figure(figsize=(20,10))
plot_tree(clf, filled=True, feature_names=selected_features, class_names=clf.classes_, rounded=True)
plt.show()
```

```python


df = merged_df[selected_features + ['GDPCAP']].copy()

df['GDPCAP'] = df['GDPCAP'].astype(float)

X = sm.add_constant(df[selected_features])
y = df['GDPCAP']

model = sm.OLS(y, X).fit()

print(model.summary())
```
